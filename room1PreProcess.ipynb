Set Geometric paramaters (Change depending on room)

```python
x_out = 6
y_out = 4;
z_out = 3;
x_in = 6;
y_in = 20;
z_in = 3;
ceiling = 3;
floor = 0;
front = 24;
back = 0;
right = 12;
left = 0;
```

Functions for the coordinate distance to each wall

```python
def dist_ceiling(row):
  return (abs(row['z'] - ceiling));
def dist_floor(row):
  return (abs(row['z'] - floor));
def dist_front(row):
  return (abs(row['y'] - front));
def dist_back(row):
  return (abs(row['y'] - back));
def dist_right(row):
  return (abs(row['x']- right));
def dist_left(row):
  return (abs(row['x']- left));
```

Function for the coordinate distance to each inlet and outlet

```python
def vect_in_x(row):
  return (row['x'] - x_in);
def vect_in_y(row):
  return (row['y'] - y_in);
def vect_in_z(row):
  return (row['z'] - z_in);
def vect_out_x(row):
  return (x_out - row['x']);
def vect_out_y(row):
  return (y_out - row['y']);
def vect_out_z(row):
  return (z_out - row['z']);
```

Functions for the absolute distance to the inlet and outlet

```python
def distance_in(row):
  return (((row['x_in'])**2+(row['y_in'])**2+(row['z_in'])**2));
def distance_out(row):
  return (((row['x_out'])**2+(row['y_out'])**2+(row['z_out'])**2));
```

Making new columns for the coordinate distance to inlet and outlet

```python
data['x_in'] = data.apply(vect_in_x, axis=1)
data['y_in'] = data.apply(vect_in_y, axis=1)
data['z_in'] = data.apply(vect_in_z, axis=1)
data['x_out'] = data.apply(vect_out_x, axis=1)
data['y_out'] = data.apply(vect_out_y, axis=1)
data['z_out'] = data.apply(vect_out_z, axis=1)
```

Making new columns for the coordinate distance to each wall

```python
data['ceiling'] = data.apply(dist_ceiling, axis=1)
data['floor'] = data.apply(dist_floor, axis=1)
data['front'] = data.apply(dist_right, axis=1)
data['back'] = data.apply(dist_left, axis=1)
data['right'] = data.apply(dist_front, axis=1)
data['left'] = data.apply(dist_back, axis=1)
```

```python
data['dist_in'] = data.apply(distance_in, axis=1)
data['dist_out'] = data.apply(distance_out, axis=1)
```

```python
data['x_in_vect'] = data['x_in'] / data['dist_in']
data['y_in_vect'] = data['y_in'] / data['dist_in']
data['z_in_vect'] = data['x_in'] / data['dist_in']
data['x_out_vect'] = data['x_out'] / data['dist_out']
data['y_out_vect'] = data['y_out'] / data['dist_out']
data['z_out_vect'] = data['z_out'] / data['dist_out']
```
