```python
import pandas as pd
```

```python
Laminar = pd.read_csv('/content/drive/MyDrive/MEng Dissertation/final/FinalModel/room2_data/FlowRegimes/laminar_data.csv')
Stagnation = pd.read_csv('/content/drive/MyDrive/MEng Dissertation/final/FinalModel/room2_data/FlowRegimes/stagnant_data.csv')
Transitional = pd.read_csv('/content/drive/MyDrive/MEng Dissertation/final/FinalModel/room2_data/FlowRegimes/transition_data.csv')
Turbulent = pd.read_csv('/content/drive/MyDrive/MEng Dissertation/final/FinalModel/room2_data/FlowRegimes/turbulent_data.csv')
```

```python
from sklearn.model_selection import train_test_split
Stagnation_train, Stagnation_test = train_test_split(Stagnation, test_size=0.2, random_state=42)
Laminar_train, Laminar_test = train_test_split(Laminar, test_size=0.2, random_state=42)
Transitional_train, Transitional_test = train_test_split(Transitional, test_size=0.2, random_state=42)
Turbulent_train, Turbulent_test = train_test_split(Turbulent, test_size=0.2, random_state=42)
```

```python
def maate(y_true, y_pred):
  absolute_difference = tf.abs(y_pred - y_true)
  epsilon = 1e-8
  absolute_actual = tf.abs(y_true) + epsilon
  ratio = absolute_difference / absolute_actual
  arctan_value = tf.atan(ratio)
  maate = tf.reduce_mean(arctan_value)
  return maate
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

input_shape = (12,)
opt = tf.keras.optimizers.Adam(learning_rate=0.0017)
model_Stagnation = Sequential([
    Dense(32, activation='relu', input_shape=input_shape),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(256, activation='relu'),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(3)  # Output layer for predicting 3 features
    ])
model_Stagnation.compile(optimizer=opt, loss = maate)

model_Stagnation.build()
model_Stagnation.summary()
```

```python
input_labels = ["r_to_outlet", "theta_to_outlet", "phi_to_outlet", "r_to_inlet", "theta_to_inlet", "phi_to_inlet", "ceiling_distance",
                    "floor_distance", "right_distance", "left_distance", "front_distance", "back_distance"]
output_labels = ["r", "theta", "phi"]
```

```python
model_Stagnation.fit(Stagnation_train[input_labels], Stagnation_train[output_labels], epochs=50, batch_size=100000)
```

```python
import tensorflow as tf

# Assuming 'model' is your trained TensorFlow model
# Save the model to a file
model_Stagnation.save('Stagnation.h5')  # or .pb for TensorFlow SavedModel format
```

```python
predictions_Stagnation = model_Stagnation.predict(Stagnation_test[input_labels], batch_size=10000)
```

```python
error_Stagnation = maate(Stagnation_test[output_labels], predictions_Stagnation).numpy()
```
