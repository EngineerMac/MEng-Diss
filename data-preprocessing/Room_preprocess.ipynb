```python
#Zero NaN values
#print(spherical_to_inlet_df.head(100))
inlet_coords = np.array([2.5, 1.25, 3])
# Calculate spherical coordinates from each row to the inlet
def cartesian_to_spherical_array_with_inlet(row, inlet_coords):
    X, Y, Z = row
    delta_x = inlet_coords[0] - X;
    delta_y = inlet_coords[1] - Y;
    delta_z = inlet_coords[2] - Z;

    r = np.sqrt(delta_x**2 + delta_y**2 + delta_z**2)+1e-8;
    theta = np.arccos(delta_z / r);

    phi = np.arctan2(delta_y,(delta_x))+(np.pi)

    return 1/r, theta, phi

# Assuming 'data' is your DataFrame

# Calculate spherical vectors to the inlet for each row
spherical_to_inlet = np.apply_along_axis(cartesian_to_spherical_array_with_inlet, 1, data[['X', 'Y', 'Z']].values, inlet_coords=inlet_coords)

# Create a DataFrame for spherical vectors to the inlet
spherical_to_inlet_df = pd.DataFrame(spherical_to_inlet, columns=['r_to_inlet', 'theta_to_inlet', 'phi_to_inlet'])
```

```python
import numpy as np
import pandas as pd

# Calculate spherical coordinates
def cartesian_to_spherical_array(cartesian_vector):
    vx, vy, vz = cartesian_vector
    r = np.sqrt(vx**2 + vy**2 + vz**2) + 1e-8
    theta = np.arccos(vz / r)
    phi = np.arctan2(vy, vx) + np.pi
    return r, theta, phi

# Convert the DataFrame to a NumPy array
data1 = np.array(data[['vx', 'vy', 'vz']])

# Apply the conversion function to the entire array
spherical_coordinates = np.apply_along_axis(cartesian_to_spherical_array, 1, data1)

# Create a DataFrame for spherical vectors
spherical_coordinates = pd.DataFrame(spherical_coordinates, columns=['r', 'theta', 'phi'])
```

```python
import numpy as np
import pandas as pd

# Calculate spherical coordinates
def cartesian_to_spherical_array(vx, vy, vz):
    r = np.sqrt(vx**2 + vy**2 + vz**2)
    theta = np.arccos(-1*vz / (r+1e-8))
    phi = np.arctan2(vy, vx)+(np.pi);
    return r, theta, phi

# Convert the DataFrame to NumPy arrays
vx_arr = data['vx'].values
vy_arr = data['vy'].values
vz_arr = data['vz'].values

# Create a progress bar
# progress_bar = tqdm(total=len(data), desc="Processing")

# Apply the conversion function to the entire arrays
spherical_coordinates = [cartesian_to_spherical_array(vx, vy, vz) for vx, vy, vz in zip(vx_arr, vy_arr, vz_arr)]
```
